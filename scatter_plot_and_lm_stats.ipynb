```python
%matplotlib inline
```

```python
import seaborn as sns
from matplotlib import pyplot
```

```python
df = sns.load_dataset("iris")
```

```python
## scatter plot
sns.regplot(x=df["sepal_length"], y=df["petal_length"])
```

```python
# adjusting graph scale
plot = sns.regplot(x=df["sepal_length"], y=df["petal_length"])
plot.set_ylim([0, 8])
```

```python
# adjusting graph style
sns.set_style("darkgrid")
sns.regplot(x=df["sepal_length"], y=df["petal_length"])
```

```python
# changing graph figure
plot = sns.regplot(x=df["sepal_length"], y=df["petal_length"])
plot.figure.set_size_inches(12, 10)
```

```python
# saving the graph image
plot = sns.regplot(x=df["sepal_length"], y=df["petal_length"])
plot.figure.savefig("lm.png")
```

```python
# regression analysis
import statsmodels.formula.api as smf
```

```python
results = smf.ols("petal_length ~ sepal_length", data=df).fit()
```

```python
results.summary()
```

```python
# multi-regression and descrete variables
results = smf.ols("petal_length ~ sepal_length + C(species)", data = df).fit()
results.summary()
```
