```python
%matplotlib inline
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
```

```python
# bar plot
sns.barplot(x=df["species"], y=df["sepal_length"])
```

```python
# FacetGrid
g = sns.FacetGrid(df, col="species")
g.map(sns.regplot, "sepal_length", "petal_length")
```

```python
g = sns.FacetGrid(df, row="species")
g.map(sns.regplot, "sepal_length", "petal_length")
```

```python
g = sns.FacetGrid(df, hue="species")
g.map(sns.regplot, "sepal_length", "petal_length")
```

```python
# lmplot
sns.lmplot("sepal_length", "petal_length", hue="species", data=df)
```

```python
# interaction 
import statsmodels.formula.api as smf
```

```python
model = smf.ols("petal_length ~ sepal_length*C(species)", df).fit()
```

```python
model.summary()
```

```python
# PairGrid
g = sns.PairGrid(df)
g = g.map(sns.regplot)
```

```python
# chage diag elements to histogram
g = sns.PairGrid(df)
g = g.map_diag(sns.distplot)
g = g.map_offdiag(sns.regplot)
```

```python
# different colors by species
from matplotlib import pyplot as plt
```

```python
g = sns.PairGrid(df, hue="species")
g = g.map_diag(plt.hist)
g = g.map_offdiag(sns.regplot)
g = g.add_legend()
```

```python
sns.pairplot(df, hue="species", kind="reg")
```
