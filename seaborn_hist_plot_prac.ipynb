```python
from matplotlib import pyplot as plt
```

```python
sns.distplot(df["sepal_length"], color="skyblue", label="Sepal Length")
sns.distplot(df["sepal_width"], color = "red", label="Sepal Width")
plt.legend()
```

```python
all_species = set(df['species'])
```

```python
colors = set(['red', 'green', 'blue'])
```

```python
list(zip(all_species, colors))
```

```python
for species, color in zip(all_species, colors):
    sns.distplot(df[df['species'] == species]["sepal_length"],
                color = color, label = species, hist= False)
plt.legend()
```
